```python
import numpy as np

# Define the conditions for each category based on the user's specifications
def categorize_prediction(row):
    if row['Is In Picture'] == 'Yes' and pd.notna(row['Gemini 1.5 Predictions']):
        return 'True Positive'
    elif row['Is In Picture'] == 'No' and pd.notna(row['Gemini 1.5 Predictions']):
        return 'False Positive'
    elif row['Is In Picture'] == 'No' and pd.isna(row['Gemini 1.5 Predictions']):
        return 'True Negative'
    elif row['Is In Picture'] == 'Yes' and pd.isna(row['Gemini 1.5 Predictions']):
        return 'False Negative'
    return 'Uncategorized'

# Apply the function to categorize each prediction
confusion_matrix_gemini['Prediction Category'] = confusion_matrix_gemini.apply(categorize_prediction, axis=1)

# Create the confusion matrix by counting each category
confusion_matrix = confusion_matrix_gemini['Prediction Category'].value_counts()

# Display the confusion matrix to the user
confusion_matrix
```

```python
# Calculate Precision, Recall, and F-1 Score based on the confusion matrix values
true_positive = confusion_matrix.get('True Positive', 0)
false_positive = confusion_matrix.get('False Positive', 0)
false_negative = confusion_matrix.get('False Negative', 0)

# Avoid division by zero in case there are no positives
precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) > 0 else 0
recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

# Display the F-1 Score along with Precision and Recall for clarity
precision, recall, f1_score
```

```python
import base64
import json
import time
from anthropic import AnthropicVertex

LOCATION = "us-east5"
client = AnthropicVertex(region=LOCATION, project_id="cocos-1682214059888")

# Local image folder path
image_folder_path = "C:\\Users\\Joon Park\\Desktop\\School\\UCHICAGO WORK\\Capstone\\Recognition_Data_Set\\"

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    image_file_name = row['Image File name']
    item_description = row['Item in image']
    image_path = f"{image_folder_path}{image_file_name}"
    image_media_type = "image/jpeg"

    # Prompt for object location
    prompt = f"""
    Please locate the object within the image based on the following description: {item_description}

    Instructions:

    	•	Horizontal Position: Select one word to describe the object’s position from left to right:
    	•	left
    	•	center
    	•	right

    	•	Vertical Position: Select one word to describe the object’s position from top to bottom:
    	•	top
    	•	center
    	•	bottom

    	•	Depth: Select one word to describe how close the object appears:
    	•	near
    	•	medium
    	•	far

     If the item is not in the image, your response should be a single word: "None'

    Your response should:

    	•	Consist of exactly three words, one from each category above.
    	•	Exclude any additional words, punctuation, or capitalization."""

    # Read and encode the image from the local file
    with open(image_path, "rb") as image_file:
        image_data = base64.b64encode(image_file.read()).decode("utf-8")

    error_count = 0
    
    # Try to send the request with error handling
    while True:
        try:
            # Create message with base64 encoded image data
            message = client.messages.create(
                max_tokens=1024,
                messages=[
                    {
                        "role": "user",
                        "content": [
                            {
                                "type": "image",
                                "source": {
                                    "type": "base64",
                                    "media_type": image_media_type,
                                    "data": image_data,
                                },
                            },
                            {
                                "type": "text",
                                "text": prompt
                            }
                        ],
                    }
                ],
                model="claude-3-haiku@20240307",
            )

            # Parse the response JSON
            json_str = message.model_dump_json(indent=2)
            message_dict = json.loads(json_str)
            extracted_text = message_dict["content"][0]["text"]

            # Print the extracted position description
            print(extracted_text)
            df.at[index, 'Claude Haiku Predictions'] = extracted_text  # Assuming response.text contains the prediction
            
            break  # Exit the loop on success

        except Exception as e:
            error_count+=1
            
            print(f"Error: {e}. Retrying in 100 seconds...")
            time.sleep(10)  # Wait 100 seconds before retrying

            if error_count >= 2:
                print(f"Skipping row {index} after 2 failed attempts.")
                break
```

```python
import string

df['True item  Location'] = df['True item  Location'].str.replace(f"[{string.punctuation}]", "", regex=True)
df.head()
```

```python
df["LR_True"] = df["True item  Location"].apply(lambda x: x.split()[0])
df["UD_True"] = df["True item  Location"].apply(lambda x: x.split()[1])
df["FN_True"] = df["True item  Location"].apply(lambda x: x.split()[2])

def recast_close_to_near(x):
    if x == 'close':
        return 'near'
    else:
        return x

df["FN_True"] = df["FN_True"].apply(lambda x: recast_close_to_near(x))
```

# Gemini

```python
prediction_column = "Gemini 1.5 Predictions"
```

```python
df["LR_Pred"] = df[prediction_column].apply(lambda x: x.split()[0] if x is not None else None)
df["UD_Pred"] = df[prediction_column].apply(lambda x: x.split()[1] if x is not None else None)
df["FN_Pred"] = df[prediction_column].apply(lambda x: x.split()[2] if x is not None else None)

accuracy_LR = (df["LR_True"] == df["LR_Pred"]).mean()
accuracy_UD = (df["UD_True"] == df["UD_Pred"]).mean()
accuracy_FN = (df["FN_True"] == df["FN_Pred"]).mean()

print(f"Accuracy for LR: {accuracy_LR}")
print(f"Accuracy for UD: {accuracy_UD}")
print(f"Accuracy for FN: {accuracy_FN}")
```
