```python
import pandas as pd
excel_filename = "SkySearch Images.xlsx"
prediction_column = "ChatGPT 4o prediction"
df = pd.read_excel(excel_filename)
df.dropna(inplace=True)
df.head()
```

```python
import string

# Remove punctuation from the 'True item Location' column
df['True item  Location'] = df['True item  Location'].str.replace(f"[{string.punctuation}]", "", regex=True)
df.head()
```

```python
df["LR_True"] = df["True item  Location"].apply(lambda x: x.split()[0])
df["UD_True"] = df["True item  Location"].apply(lambda x: x.split()[1])
df["FN_True"] = df["True item  Location"].apply(lambda x: x.split()[2])
def recast_close_to_near(x):
    if x == 'close':
        return 'near'
    else:
        return x

df["FN_True"] = df["FN_True"].apply(lambda x: recast_close_to_near(x))

df["LR_Pred"] = df[prediction_column].apply(lambda x: x.split()[0])
df["UD_Pred"] = df[prediction_column].apply(lambda x: x.split()[1])
df["FN_Pred"] = df[prediction_column].apply(lambda x: x.split()[2])

accuracy_LR = (df["LR_True"] == df["LR_Pred"]).mean()
accuracy_UD = (df["UD_True"] == df["UD_Pred"]).mean()
accuracy_FN = (df["FN_True"] == df["FN_Pred"]).mean()

print(f"Accuracy for LR: {accuracy_LR}")
print(f"Accuracy for UD: {accuracy_UD}")
print(f"Accuracy for FN: {accuracy_FN}")
```
